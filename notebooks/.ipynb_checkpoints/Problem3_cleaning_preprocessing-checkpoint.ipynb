```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
from scipy import stats

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
asthma_dataset = pd.read_csv("../data/asthma_disease_data.csv")
```

```python
asthma_dataset.dtypes
```

```python
asthma_dataset.describe(include="all")
```

```python
asthma_dataset.isnull().sum()
```

```python
asthma_dataset = asthma_dataset.dropna()
```

```python
asthma_dataset
```

```python
# Use box plots to visualize and detect outliers.

sns.boxplot(x=asthma_dataset["LungFunctionFEV1"])
plt.show()
```

```python
# measures how many standard deviations a data point is from the mean of the dataset

z_scores = np.abs(stats.zscore(asthma_dataset["Wheezing"]))
asthma_dataset_no_outliers = asthma_dataset[(z_scores < 3)]
```

```python
z_scores
```

```python
asthma_dataset_no_outliers
```

```python
# IQR (Interquartile Range) - measures the spread of the middle 50% of a dataset:

Q1 = asthma_dataset_no_outliers["Wheezing"].quantile(0.25)
Q3 = asthma_dataset_no_outliers["Wheezing"].quantile(0.75)
IQR = Q3 - Q1
asthma_dataset_no_outliers = asthma_dataset[~((asthma_dataset["Wheezing"] < (Q1 - 1.5 * IQR)) |asthma_dataset["Wheezing"] > (Q3 + 1.5 * IQR))]
```

```python
Q1
```

```python
Q3
```

```python
IQR
```

```python
asthma_dataset_no_outliers
```

```python
# Convert continuous variables into categorical variables by binning

asthma_dataset_no_outliers["BMI"] = pd.cut(asthma_dataset_no_outliers["BMI"], bins=5)
```

```python
asthma_dataset_no_outliers["BMI"]
```

```python
# Convert categorical variables into numerical format.

le = LabelEncoder()
asthma_dataset_no_outliers["DoctorInCharge"] = le.fit_transform(asthma_dataset_no_outliers["DoctorInCharge"])
```

```python
asthma_dataset_no_outliers["DoctorInCharge"]
```

```python
# Standardization (Z-Score Scaling):

scaler = StandardScaler()
asthma_dataset_no_outliers[["LungFunctionFVC"]] = scaler.fit_transform(asthma_dataset_no_outliers[["LungFunctionFVC"]])
```

```python
asthma_dataset_no_outliers[["LungFunctionFVC"]]
```

```python
# Normalization (Min-Max Scaling):

scaler = MinMaxScaler()
asthma_dataset_no_outliers[["LungFunctionFEV1"]] = scaler.fit_transform(asthma_dataset_no_outliers[["LungFunctionFEV1"]])
```

```python
asthma_dataset_no_outliers[["LungFunctionFEV1"]]
```

```python
asthma_dataset_no_outliers.to_csv("../data/cleaned_asthma_dataset.csv", index=False)
```
