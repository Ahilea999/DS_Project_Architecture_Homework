```python
def convert_interval_to_midpoint(val):
    if isinstance(val, pd._libs.interval.Interval):
        return val.mid 
    else:
        return val
```

```python

asthma_dataset_no_outliers = asthma_dataset_no_outliers.applymap(convert_interval_to_midpoint)
```

* Data transforming and scaling

```python
# Continuous features that need scaling

continuous_features = ["LungFunctionFVC", "LungFunctionFEV1", "DustExposure"]
```

```python
# Initialize the scaler

scaler = StandardScaler()
```

```python
# Scale continuous features

asthma_dataset_no_outliers[continuous_features] = scaler.fit_transform(asthma_dataset_no_outliers[continuous_features])
```

```python
asthma_dataset_no_outliers[continuous_features]
```

* Feature Engineering

```python
asthma_dataset_no_outliers["FVC_Wheezing"] = asthma_dataset_no_outliers["LungFunctionFVC"] * asthma_dataset_no_outliers["Wheezing"]
asthma_dataset_no_outliers["FEV1_Coughing"] = asthma_dataset_no_outliers["LungFunctionFEV1"] * asthma_dataset_no_outliers["Coughing"]
asthma_dataset_no_outliers["Dust_NighttimeSymptoms"] = asthma_dataset_no_outliers["DustExposure"] * asthma_dataset_no_outliers["NighttimeSymptoms"]
```
