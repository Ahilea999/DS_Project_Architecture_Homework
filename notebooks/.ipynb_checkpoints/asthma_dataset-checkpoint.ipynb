```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
asthma_dataset = pd.read_csv("data/asthma_disease_data.csv")
```

```python
asthma_dataset
```

```python
# Check for missing values
# The result show no missing values

asthma_dataset.isnull().sum()
```

```python
# Check the data types of each column

asthma_dataset.dtypes
```

```python
# Get summary statistics for numeric columns

asthma_dataset.describe()
```

```python
# Plot distribution of a key variable (in this case "Age")

sns.histplot(asthma_dataset["Age"], kde=True)
plt.title("Distribution of the disease according to the age")
plt.show()
```

```python
asthma_dataset["DoctorInCharge"].value_counts()
```

```python
asthma_dataset = asthma_dataset.drop(columns=["DoctorInCharge"])
```

```python
# Correlation matrix

correlation_matrix = asthma_dataset.corr()
```

```python
# Plotting the correlation matrix

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Count plot of a categorical variable (e.g., Gender)

sns.countplot(x="Gender", data=asthma_dataset)
plt.title("Distribution of Gender")
plt.show()
```

```python
# Box plot of Coughing dependent by diet quality

sns.boxplot(x="Coughing", y="DietQuality", data=asthma_dataset)
plt.title("Asthma Severity by Diet quality")
plt.show()
```
